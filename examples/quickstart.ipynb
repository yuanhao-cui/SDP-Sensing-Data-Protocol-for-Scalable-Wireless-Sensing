# WSDP Quickstart

This notebook walks through the basic workflow of the **WSDP** (Wi-Fi Sensing Data Protocol) library:

1. Install the package
2. Download a sample dataset
3. Run the classification pipeline
4. Explore the model registry
5. Use algorithm / preprocessing registries

```python
# Install wsdp (uncomment if running in Colab / fresh environment)
# !pip install wsdp
```

```python
import wsdp

print(f"wsdp version: {wsdp.__version__}")

# Download a sample dataset
wsdp.download(
    dataset="UT_HAR",
    output_folder="./data",
)
```

```python
# Run the full classification pipeline
wsdp.pipeline(
    input_path="./data",
    output_folder="./output",
    dataset="UT_HAR",
    num_seeds=1,        # use 1 seed for a quick demo
    num_epochs=5,       # few epochs for demonstration
)
```

## Available Models

WSDP ships with a rich model registry spanning baselines, mainstream architectures,
state-of-the-art models, and specialised WiFi CSI models.

```python
from wsdp.models import list_models

# List every registered model
for name in list_models():
    print(name)
```

## Custom Preprocessing

You can inspect and extend the algorithm / processor registries to plug in
your own preprocessing steps.

```python
from wsdp.processors.base_processor import BaseProcessor

# Inspect the default processing pipeline
processor = BaseProcessor()
print(f"Processor: {processor}")

# You can also use the algorithm registry directly:
from wsdp.processors import ALGORITHM_REGISTRY

print("\nRegistered algorithms:")
for name in sorted(ALGORITHM_REGISTRY):
    print(f"  - {name}")
```
